```python
# Membuat data deret waktu dengan timestep
timesteps = 5  # Menggunakan data 5 hari sebelumnya

def create_sequences(X, y, timesteps):
    X_seq, y_seq = [], []
    for i in range(len(X) - timesteps):
        X_seq.append(X[i:i+timesteps])
        y_seq.append(y[i+timesteps])
    return np.array(X_seq), np.array(y_seq)

X_seq, y_seq = create_sequences(X_scaled, y_scaled, timesteps)
```

```python
from keras_tuner import RandomSearch
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Fungsi untuk membangun model
def build_model(hp):
    model = Sequential()
    # Menentukan jumlah unit LSTM
    model.add(LSTM(
        units=hp.Int('units', min_value=32, max_value=128, step=32),
        activation='tanh',
        input_shape=(X_train.shape[1], X_train.shape[2])
    ))
    model.add(Dense(1))  # Output layer
    model.compile(
        optimizer='adam',
        loss='mse',
        metrics=['mae']
    )
    return model

# Random Search
tuner = RandomSearch(
    build_model,
    objective='val_mae',
    max_trials=10,
    executions_per_trial=1,
    directory='lstm_tuning',
    project_name='weather_prediction'
)

# Menjalankan Tuning
tuner.search(X_train, y_train, epochs=10, validation_split=0.2)

# Menampilkan Hyperparameter Terbaik
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"Best number of units: {best_hps.get('units')}")
```

```python
import matplotlib.pyplot as plt

# Prediksi pada data uji
y_pred = model.predict(X_test)
y_pred = y_pred.flatten()  # Pastikan y_pred 1 dimensi

# Plot hasil prediksi vs data asli
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Actual Data', color='blue')  # Data asli
plt.plot(y_pred, label='Predicted Data', color='red')  # Hasil prediksi model
plt.title('Predicted vs Actual Values')
plt.xlabel('Time Step')
plt.ylabel('Temperature (or target variable)')
plt.legend()
plt.grid(True)  # Tambahkan grid untuk tampilan lebih jelas
plt.show()
```
